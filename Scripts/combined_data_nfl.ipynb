```python
import os
import pandas as pd
import numpy as np
```

```python
input_file_draft = os.path.join("..","Clean_Data","draft.csv")
input_file_fumbl = os.path.join("..","Clean_Data","fumblForced.csv")
input_file_games = os.path.join("..","Clean_Data","games.csv")
input_file_interceptions = os.path.join("..","Clean_Data","interceptions.csv")
input_file_players = os.path.join("..","Clean_Data","players.csv")
input_file_plays = os.path.join("..","Clean_Data","plays.csv")
input_file_sacks = os.path.join("..","Clean_Data","sacks.csv")

output_file_int = os.path.join("..","Clean_Data","int_combined.csv")
output_file_sacks = os.path.join("..","Clean_Data","sacks_combined.csv")
output_file_fumbl = os.path.join("..","Clean_Data","fumbl_combined.csv")
output_file_nfl = os.path.join("..","Clean_Data","nlf_combined.csv")
```

```python
draft_df = pd.read_csv(input_file_draft)
draft_df = draft_df.rename(columns={'teamId':'DraftTeamId'})
draft_df.head(200)
```

```python
fumbl_df = pd.read_csv(input_file_fumbl)
fumbl_df = fumbl_df.rename(columns={'teamId':'FumblTeamId'})
fumbl_df.head()
```

```python
games_df = pd.read_csv(input_file_games)
games_df.head()
```

```python
interceptions_df = pd.read_csv(input_file_interceptions)
interceptions_df = interceptions_df.rename(columns={'teamId':'IntTeamId'})
interceptions_df.head()
```

```python
players_df = pd.read_csv(input_file_players)
players_df
# players_df.loc[players_df['playerId']==20100001]
```

```python
sacks_df = pd.read_csv(input_file_sacks)
sacks_df = sacks_df.rename(columns={'teamId':'SacksTeamId'})
sacks_df.head()
```

```python
plays_df = pd.read_csv(input_file_plays)
plays_df.head(200)
```

```python
int_merged_df = pd.DataFrame(pd.merge(players_df,interceptions_df,how='left',on='playerId'))
int_merged_df = pd.merge(int_merged_df,plays_df,how='left',on='playId')
int_merged_df = pd.merge(int_merged_df,games_df,how='left',on='gameId')
int_merged_df = pd.merge(int_merged_df,draft_df,how='left',on='playerId')
int_merged_df = int_merged_df.dropna().drop_duplicates(subset='playId')
display(int_merged_df)
int_merged_df.count()
```

```python
fumbl_merged_df = pd.DataFrame(pd.merge(players_df,fumbl_df,how='left',on='playerId'))
fumbl_merged_df = pd.merge(fumbl_merged_df,plays_df,how='left',on='playId')
fumbl_merged_df = pd.merge(fumbl_merged_df,games_df,how='left',on='gameId')
fumbl_merged_df = pd.merge(fumbl_merged_df,draft_df,how='left',on='playerId')
fumbl_merged_df = fumbl_merged_df.dropna().drop_duplicates(subset='playId')
display(fumbl_merged_df)
fumbl_merged_df.count()
```

```python
sacks_merged_df = pd.DataFrame(pd.merge(players_df,sacks_df,how='left',on='playerId'))
sacks_merged_df = pd.merge(sacks_merged_df,plays_df,how='left',on='playId')
sacks_merged_df = pd.merge(sacks_merged_df,games_df,how='left',on='gameId')
sacks_merged_df = pd.merge(sacks_merged_df,draft_df,how='left',on='playerId')
sacks_merged_df = sacks_merged_df.dropna().drop_duplicates(subset='playId')
display(sacks_merged_df)
sacks_merged_df.count()
```

```python
nfl_merged_df = pd.concat([int_merged_df,sacks_merged_df], sort = False)
nfl_merged_df = pd.concat([nfl_merged_df,fumbl_merged_df], sort = False)
nfl_merged_df = nfl_merged_df.loc[nfl_merged_df["season"] >= 2010]
display(nfl_merged_df)
nfl_merged_df.count()
```

```python
col = list(nfl_merged_df.columns)
for each_col in col:
    null_value = len(nfl_merged_df.loc[nfl_merged_df[each_col].isnull()])
    print(f"Filas vacias en {each_col}: {null_value}")
```

```python
int_merged_df.to_csv(output_file_int,index=False,header=True)
sacks_merged_df.to_csv(output_file_sacks,index=False,header=True)
fumbl_merged_df.to_csv(output_file_fumbl,index=False,header=True)
nfl_merged_df.to_csv(output_file_nfl,index=False,header=True)
```
