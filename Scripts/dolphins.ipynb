```python
import os
import pandas as pd
import numpy as np
```

```python
input_file_draft = os.path.join("..","..","Resources","draft.csv")

output_file_dolphins = os.path.join("..","Clean_Data","dolphins.csv")
```

```python
draft_df = pd.read_csv(input_file_draft)
dolphins = draft_df[['playerId','draft','teamId','draftTeam','nameFull','position']]
dolphins
```

```python
dolphins=dolphins[dolphins['draft']>=2014]
dolphins=dolphins[dolphins['draftTeam']=='MIA']
dolphins
```

```python
dolphins.to_csv(output_file_dolphins,index=False,header=True)
```
