```python
import os
import pandas as pd
import numpy as np
```

```python
input_file = os.path.join("..","Resources")
games_file = 'games.csv'
plays_file = 'plays.csv'
sacks_file = 'sacks_original.csv'

output_file = os.path.join('..','Output','sacks_combined.csv')
```

```python
games_df = pd.read_csv(input_file + '/' + games_file)

games_df = games_df[['gameId','season','week','gameDate','homeTeamId','visitorTeamId','seasonType','winningTeam']]
games_df = games_df.loc[games_df['season']>=2014]
games_df
```

```python
sacks_df = pd.read_csv(input_file + '/' + sacks_file)
sacks_df = sacks_df.loc[sacks_df['sackType']==1]
sacks_df = sacks_df.loc[sacks_df['sackNull']==0]
sacks_df = sacks_df[['playId','playerId']]
sacks_df
```

```python
plays_df = pd.read_csv(input_file + '/' + plays_file)
plays_df = plays_df[['playId','gameId']]
plays_df
```

```python
plays_x_game = pd.merge(plays_df,games_df,how='inner',on='gameId')
plays_x_game
```

```python
sacks_x_play = pd.merge(sacks_df,plays_x_game,how='inner',on='playId')
sacks_x_play
```

```python
sacks_x_play.to_csv(output_file,index=False,header=True)
```
