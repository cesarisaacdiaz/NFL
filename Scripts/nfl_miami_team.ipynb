```python
import os
import pandas as pd
import numpy as np
```

```python
input_file = os.path.join("..","..","Resources")
draft_file = 'draft.csv'
players_file = 'players.csv'
game_participation_file = 'gameParticipation.csv'
games_file = 'games.csv'

fumble_file = 'fumbles.csv'
interceptions_file = 'interceptions.csv'
sacks_file = 'sacks.csv'

output_file = os.path.join('..','Clean_Data','total_players.csv')
```

```python
draft_df = pd.read_csv(input_file +'/'+ draft_file)
draft_df.head()

#draft_df = draft_df.loc[draft_df['teamId']==2700]
team_df = draft_df[['teamId','draftTeam']]
team_df = team_df.drop_duplicates()
team_df
```

```python
players_df = pd.read_csv(input_file + '/' + players_file)
players_df.head()

players_df = players_df[['playerId','nameFull','position']]
players_df.head()
```

```python
participation_df = pd.read_csv(input_file + '/'+ game_participation_file)
participation_df.head()

participation_df = participation_df[['gameId','teamId','playerId','nameFull','position','gamePartUnit']]
#participation_df['gamePartUnit'].value_counts()
#miami_team_df = participation_df.loc[participation_df['teamId']==2700]
defenders_team_df = participation_df.loc[participation_df['gamePartUnit']=='defense']
defenders_team_df
```

```python
defenders_team_df = pd.merge(defenders_team_df,team_df,how='inner',on='teamId')
defenders_team_df
```

```python
games_df = pd.read_csv(input_file + '/'+ games_file)
games_df.head()

games_df = games_df[['gameId','season','week','gameDate','homeTeamId','visitorTeamId','seasonType','winningTeam']]
games_df = games_df.loc[games_df['season']>=2014]
#season_2019 = season_2019.loc[season_2019['seasonType']!='PRE']
#season_2019 = season_2019[(season_2019['homeTeamId']==2700)|(season_2019['homeTeamId']==2700)]
games_df.head()
```

```python
players_final_df = pd.merge(games_df,defenders_team_df,how='left',on='gameId')
players_final_df.head()

players_final_df = players_final_df[['playerId','nameFull','position','draftTeam']]
players_final_df = players_final_df.drop_duplicates()
players_final_df
```

```python
defenders_team_df
players_group = defenders_team_df.groupby(by='playerId')
games_played = pd.DataFrame(players_group['gameId'].count())
games_played = games_played.reset_index()
games_played
```

```python
players_final_df = pd.merge(players_final_df,games_played,how='left',on='playerId')
players_final_df
```

```python
fumble_df = pd.read_csv(input_file +'/'+ fumble_file)
#fumble_df = fumble_df[['playerId','playId','gameId']]
fumble_df.head()

fumble_group = fumble_df.groupby(by='playerId')
count_fumbled = pd.DataFrame(fumble_group['playerId'].count())
count_fumbled = count_fumbled.rename(columns = {'playerId':'fumbles'})
count_fumbled = count_fumbled.reset_index()
count_fumbled
```

```python
players_final_df = pd.merge(players_final_df,count_fumbled,how='left',on='playerId')
players_final_df
```

```python
interceptions_df = pd.read_csv(input_file +'/'+ interceptions_file)

#interceptions_df = interceptions_df[['playerId','playId','gameId']]
interceptions_df.head()

interceptions_group = interceptions_df.groupby(by='playerId')
count_interceptions = pd.DataFrame(interceptions_group['playerId'].count())
count_interceptions = count_interceptions.rename(columns = {'playerId':'interceptions'})
count_interceptions = count_interceptions.reset_index()
count_interceptions
```

```python
players_final_df = pd.merge(players_final_df,count_interceptions,how='left',on='playerId')
players_final_df
```

```python
sacks_df = pd.read_csv(input_file +'/'+ sacks_file)

#sacks_df = sacks_df[['playerId','playId','gameId']]
sacks_df.head()

sacks_group = sacks_df.groupby(by='playerId')
count_sacks = pd.DataFrame(sacks_group['playerId'].count())
count_sacks = count_sacks.rename(columns = {'playerId':'sacks'})
count_sacks = count_sacks.reset_index()
count_sacks
```

```python
players_final_df = pd.merge(players_final_df,count_sacks,how='left',on='playerId')
players_final_df
```
