```python
import pandas as pd
```

```python
#Path of the files that we want to open
input_path = '../Resources/'
output_path = '../Clean_Data/'

fumbl_forced_file = 'fumblForced.csv'
plays_file = 'plays.csv'

#Function to open files using pandas
def open_file(path_file):
    df = pd.read_csv(path_file)
    return df

#Save the df in variables
fumbl_forced_df = open_file(input_path + fumbl_forced_file)
plays_df = open_file(input_path + plays_file)
```

```python
#Display the tables 
display(fumbl_forced_df.head())
display(plays_df.head())
```

```python
#Choose the columns that we want
plays_clean = plays_df[['playId','gameId']]
#Print null values for each column
print('plays.csv\n')
print(f'Filas vacias en "playid": {len(plays_clean.loc[plays_clean["playId"].isnull()])}')
print(f'Filas vacias en "gameId": {len(plays_clean.loc[plays_clean["gameId"].isnull()])}')
print('\n-----------------------------\n')

      
#Choose the columns that we want
fumbl_forced_clean = fumbl_forced_df[['playId','teamId','playerId','fumForcedPosition','fumForcedTurnover',
                                      'fumForcedNull']]
#Store columns in a list
col = list(fumbl_forced_clean.columns)
#Print null values for each column
print('fumblForced.csv.csv\n')
for each_col in col:
      null_value = len(fumbl_forced_clean.loc[fumbl_forced_clean[each_col].isnull()])
      print(f'Filas vacias en {each_col}: {null_value}')
```

```python
#Drop null values
fumbl_forced_clean = fumbl_forced_clean.dropna()
```

```python
#Show data type of each column
display(fumbl_forced_clean.dtypes)
display(plays_clean.dtypes)
```

```python
#From 'fumbl_forced_clean' table convert playerId to Integers
fumbl_forced_clean['playerId'] = fumbl_forced_clean['playerId'].astype('int')
fumbl_forced_clean['fumForcedTurnover'] = fumbl_forced_clean['fumForcedTurnover'].astype('int')

#Display again the data type of each column in 'fumbl_forced_clean' table
fumbl_forced_clean.dtypes
```

```python
#Show the Data cleaned
display(fumbl_forced_clean.head())
display(plays_clean.head())
```

```python
#Save cleaned data in .csv files

fumbl_forced_clean.to_csv((output_path + fumbl_forced_file), index=False, header=True)

plays_clean.to_csv((output_path + plays_file), index=False, header=True)
```
