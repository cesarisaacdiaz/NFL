```python
#Dependencies
import pandas as pd
```

```python
# File to Load 
file_to_load = "../Resources/interceptions.csv"

# Read Purchasing File and store into Pandas data frame
interceptions_data = pd.read_csv(file_to_load)
interceptions_data
```

```python
#Cleaning for empty values
cleaned_data = interceptions_data.dropna(axis=0)

#Remove useless columns
narrow_data = cleaned_data.drop(columns=['interceptionId','teamId'])

#Filter for specific value
narrow_data = narrow_data[narrow_data['int']==1]

#Filter for specific value
narrow_data = narrow_data[narrow_data['intNull']==0]

narrow_data
```

```python
#Save dataframe to csv
narrow_data.to_csv("../Clean_Data/interceptions.csv", index=False, header=True)
```
