```python
%matplotlib notebook
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_path = os.path.join("..","Clean_Data")
output_path = os.path.join("..","Graphs")
```

```python
complete_df = pd.read_csv(input_path +'/total_players.csv')
display(complete_df)
```

```python

team='MIA'
position_graph = complete_df[complete_df['draftTeam']==team]
# display(position_graph)
position_graph= position_graph.groupby('position').sum().reset_index()
# display(position_graph)
position_graph= position_graph[['position','gameId','fumbles','interceptions','sacks']]
# display(position_graph)
position_graph2 = complete_df[complete_df['draftTeam']==team].groupby('position').count().reset_index()
# display(position_graph2)
position_graph = pd.DataFrame.merge(position_graph,position_graph2[['position','playerId']])
# display(position_graph)
position_graph = pd.DataFrame.merge(position_graph,complete_df[['position','#players-pos-league','#games-pos-league','#fumbles-pos-league','#int-pos-league','#sacks-pos-league']],how='left',on='position').drop_duplicates()
# display(position_graph)
position_graph = position_graph.rename(columns={'gameId':f'games-pos-{team}','playerId':f'players-pos-{team}','fumbles':f'fumbles-{team}','interceptions':f'interceptions-{team}','sacks':f'sacks_{team}'})
display(position_graph)
```

```python
position_graph[f'avg-fumbles-{team}'] = position_graph[f'fumbles-{team}']/position_graph[f'players-pos-{team}']/position_graph[f'games-pos-{team}']
position_graph[f'avg-interceptions-{team}'] = position_graph[f'interceptions-{team}']/position_graph[f'players-pos-{team}']/position_graph[f'games-pos-{team}']
position_graph[f'avg-sacks-{team}'] = position_graph[f'sacks_{team}']/position_graph[f'players-pos-{team}']/position_graph[f'games-pos-{team}']
position_graph['avg-fumbles-league'] = position_graph['#fumbles-pos-league']/position_graph['#players-pos-league']/position_graph[f'#games-pos-league']
position_graph['avg-interceptions-league'] = position_graph['#int-pos-league']/position_graph['#players-pos-league']/position_graph[f'#games-pos-league']
position_graph['avg-sacks-league'] = position_graph['#sacks-pos-league']/position_graph['#players-pos-league']/position_graph[f'#games-pos-league']
position_graph = position_graph[['position',f'avg-fumbles-{team}',f'avg-interceptions-{team}',f'avg-sacks-{team}','avg-fumbles-league','avg-interceptions-league','avg-sacks-league']]
display(position_graph)
```

```python
fig, ax = plt.subplots()
x = np.arange(len(position_graph['position']))
p1 = ax.bar(x, position_graph[f'avg-interceptions-{team}'], width=0.4,color='orange')
p2 = ax.bar(x+0.4, position_graph['avg-interceptions-league'], width=0.4,color='b')
ax.set_title('Average Interceptions')
ax.set_xticks(x + 0.4 / 2)
ax.set_xticklabels(position_graph['position'])
ax.legend((p1[0], p2[0]), (team, 'League'))
ax.set_ylabel('Interceptions')
ax.autoscale_view()
fig.savefig(output_path+f"/{team}_positions_vs_league_positions_interceptions")
plt.show()
```

```python
fig, ax = plt.subplots()
x = np.arange(len(position_graph['position']))
p1 = ax.bar(x-0.2, position_graph[f'avg-fumbles-{team}'], width=0.4,color='orange')
p2 = ax.bar(x + 0.2, position_graph['avg-fumbles-league'], width=0.4,color='b')
ax.set_title('Average Fumbles')
ax.set_xticks(x + 0.4 / 2)
ax.set_xticklabels(position_graph['position'])
ax.legend((p1[0], p2[0]), (team, 'League'))
ax.set_ylabel('Fumbles')
ax.autoscale_view()
fig.savefig(output_path+f"/{team}_positions_vs_league_positions_fumbles")
plt.show()
```

```python
fig, ax = plt.subplots()
x = np.arange(len(position_graph['position']))
p1 = ax.bar(x-0.2, position_graph[f'avg-sacks-{team}'], width=0.4,color='orange')
p2 = ax.bar(x + 0.2, position_graph['avg-sacks-league'], width=0.4,color='b')
ax.set_title('Average Sacks')
ax.set_xticks(x + 0.4 / 2)
ax.set_xticklabels(position_graph['position'])
ax.legend((p1[0], p2[0]), (team, 'League'))
ax.set_ylabel('Sacks')
ax.set_xlabel('Positions')
ax.autoscale_view()
fig.savefig(output_path+f"/{team}_positions_vs_league_positions_sacks")
plt.show()
```

```python
players_stats = complete_df[['playerId','nameFull','position','draftTeam','gameId','fumbles','interceptions','sacks']]
players_stats['avg_player_fumbles']=players_stats['fumbles']/players_stats['gameId']
players_stats['avg_player_interceptions']=players_stats['interceptions']/players_stats['gameId']
players_stats['avg_player_sacks']=players_stats['sacks']/players_stats['gameId']
players_stats = players_stats[['playerId','nameFull','position','draftTeam','gameId','avg_player_fumbles','avg_player_interceptions','avg_player_sacks']]
players_stats = players_stats.rename(columns={'gameId':'games_played','draftTeam':'Team'})
display(players_stats)
```

```python
by_team_graphs = players_stats[players_stats['games_played']>=20]
by_team_graphs = by_team_graphs.groupby('Team').mean().reset_index()
by_team_graphs = by_team_graphs[['Team','avg_player_fumbles','avg_player_interceptions','avg_player_sacks']]
display(by_team_graphs)
avg_league_interceptions = by_team_graphs['avg_player_interceptions'].sum()/by_team_graphs['Team'].count()
avg_league_fumbles = by_team_graphs['avg_player_fumbles'].sum()/by_team_graphs['Team'].count()
avg_league_sacks = by_team_graphs['avg_player_sacks'].sum()/by_team_graphs['Team'].count()
by_team_graphs['avg_league_fumbles'] = avg_league_fumbles
by_team_graphs['avg_league_interceptions'] = avg_league_interceptions
by_team_graphs['avg_league_sacks'] = avg_league_sacks
by_team_graphs['total_avg_plays'] = by_team_graphs['avg_player_fumbles'] + by_team_graphs['avg_player_interceptions'] + by_team_graphs['avg_player_sacks']
display(by_team_graphs)
```

```python
width=0.5
by_team_graphs[['avg_player_interceptions']].plot(kind='bar', width =width)
by_team_graphs['avg_league_interceptions'].plot()

ax = plt.gca()
plt.xlim([-width, len(by_team_graphs['Team'])-width])
ax.set_xticklabels(by_team_graphs['Team'],rotation=90)
plt.show()
```

```python
width=0.5
by_team_graphs[['avg_player_fumbles']].plot(kind='bar', width =width)
by_team_graphs['avg_league_fumbles'].plot()

ax = plt.gca()
plt.xlim([-width, len(by_team_graphs['Team'])-width])
ax.set_xticklabels(by_team_graphs['Team'],rotation=90)
plt.show()
```

```python
width=0.5
by_team_graphs[['avg_player_sacks']].plot(kind='bar', width =width)
by_team_graphs['avg_league_sacks'].plot()

ax = plt.gca()
plt.xlim([-width, len(by_team_graphs['Team'])-width])
ax.set_xticklabels(by_team_graphs['Team'],rotation=90)
plt.show()
```

```python
# I M P O R T A N T E: Se incluye un filtro para que solo aparezcan jugadores con más de 20 juegos
# porque con pocos juegos algunos tienen estadisticas muy altas
# solo es un número que a mi se me ocurrió, no está basado en nada y puede cambiarse o eliminarse
by_player_graphs = players_stats[players_stats['games_played']>=20]
avg_league_interceptions = by_player_graphs['avg_player_interceptions'].sum()/by_player_graphs['playerId'].count()
avg_league_fumbles = by_player_graphs['avg_player_fumbles'].sum()/by_player_graphs['playerId'].count()
avg_league_sacks = by_player_graphs['avg_player_sacks'].sum()/by_player_graphs['playerId'].count()
by_player_graphs['avg_league_fumbles'] = avg_league_fumbles
by_player_graphs['avg_league_interceptions'] = avg_league_interceptions
by_player_graphs['avg_league_sacks'] = avg_league_sacks
display(by_player_graphs.sort_values(['avg_player_sacks'],ascending=False))
```

```python
width=0.5
by_player_graphs[['avg_player_interceptions']].plot(kind='bar', width =width)
by_player_graphs['avg_league_interceptions'].plot()

ax = plt.gca()
plt.xlim([-width, len(by_player_graphs['nameFull'])-width])
ax.set_xticklabels(by_player_graphs['nameFull'],rotation=90)
fig = plt.figure()
fig.set_figheight(10)
fig.set_figwidth(1000)
plt.show()
```
