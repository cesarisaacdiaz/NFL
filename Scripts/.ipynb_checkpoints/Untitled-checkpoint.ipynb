```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = os.path.join("..","Clean_Data")
```

```python
complete_df = pd.read_csv(path +'/total_players.csv')
display(complete_df)
```

```python

team='MIA'
position_graph = complete_df[complete_df['draftTeam']==team]
# display(position_graph)
position_graph= position_graph.groupby('position').sum().reset_index()
# display(position_graph)
position_graph= position_graph[['position','gameId','fumbles','interceptions','sacks']]
# display(position_graph)
position_graph2 = complete_df[complete_df['draftTeam']==team].groupby('position').count().reset_index()
# display(position_graph2)
position_graph = pd.DataFrame.merge(position_graph,position_graph2[['position','playerId']])
# display(position_graph)
position_graph = pd.DataFrame.merge(position_graph,complete_df[['position','#players-pos-league','#games-pos-league','#fumbles-pos-league','#int-pos-league','#sacks-pos-league']],how='left',on='position').drop_duplicates()
# display(position_graph)
position_graph = position_graph.rename(columns={'gameId':f'games-pos-{team}','playerId':f'players-pos-{team}','fumbles':f'fumbles-{team}','interceptions':f'interceptions-{team}','sacks':f'sacks_{team}'})
display(position_graph)
```

```python
position_graph[f'avg-fumbles-{team}'] = position_graph[f'fumbles-{team}']/position_graph[f'players-pos-{team}']
position_graph[f'avg-interceptions-{team}'] = position_graph[f'interceptions-{team}']/position_graph[f'players-pos-{team}']
position_graph[f'avg-sacks-{team}'] = position_graph[f'sacks_{team}']/position_graph[f'players-pos-{team}']
position_graph['avg-fumbles-league'] = position_graph['#fumbles-pos-league']/position_graph['#players-pos-league']
position_graph['avg-interceptions-league'] = position_graph['#int-pos-league']/position_graph['#players-pos-league']
position_graph['avg-sacks-league'] = position_graph['#sacks-pos-league']/position_graph['#players-pos-league']
position_graph = position_graph[['position',f'avg-fumbles-{team}',f'avg-interceptions-{team}',f'avg-sacks-{team}','avg-fumbles-league','avg-interceptions-league','avg-sacks-league']]
display(position_graph)
```

```python
fig, ax = plt.subplots()
x = np.arange(len(position_graph['position']))
p1 = ax.bar(x, position_graph[f'avg-interceptions-{team}'], width=0.4,color='orange')
p2 = ax.bar(x+0.4, position_graph['avg-interceptions-league'], width=0.4,color='b')
ax.set_title('Average Interceptions')
ax.set_xticks(x + 0.4 / 2)
ax.set_xticklabels(position_graph['position'])
ax.legend((p1[0], p2[0]), (team, 'League'))
ax.set_ylabel('Interceptions')
ax.autoscale_view()
plt.show()
```

```python
fig, ax = plt.subplots()
x = np.arange(len(position_graph['position']))
p1 = ax.bar(x-0.2, position_graph[f'avg-fumbles-{team}'], width=0.4,color='orange')
p2 = ax.bar(x + 0.2, position_graph['avg-fumbles-league'], width=0.4,color='b')
ax.set_title('Average Fumbles')
ax.set_xticks(x + 0.4 / 2)
ax.set_xticklabels(position_graph['position'])
ax.legend((p1[0], p2[0]), (team, 'League'))
ax.set_ylabel('Fumbles')
ax.autoscale_view()
plt.show()
```

```python
fig, ax = plt.subplots()
x = np.arange(len(position_graph['position']))
p1 = ax.bar(x-0.2, position_graph[f'avg-sacks-{team}'], width=0.4,color='orange')
p2 = ax.bar(x + 0.2, position_graph['avg-sacks-league'], width=0.4,color='b')
ax.set_title('Average Sacks')
ax.set_xticks(x + 0.4 / 2)
ax.set_xticklabels(position_graph['position'])
ax.legend((p1[0], p2[0]), (team, 'League'))
ax.set_ylabel('Sacks')
ax.autoscale_view()
plt.show()
```
