```python
import os
import pandas as pd
import numpy as np
```

```python
input_file = os.path.join("..","Resources")
games_file = 'games.csv'
plays_file = 'plays.csv'
interceptions_file = 'interceptions.csv'

output_file = os.path.join('..','Output','interceptions_combined.csv')
```

```python
games_df = pd.read_csv(input_file + '/' + games_file)

games_df = games_df[['gameId','season','week','gameDate','homeTeamId','visitorTeamId','seasonType','winningTeam']]
games_df = games_df.loc[games_df['season']>=2014]
games_df
```

```python
interceptions_df = pd.read_csv(input_file + '/' + interceptions_file)
interceptions_df = interceptions_df.loc[interceptions_df['int']==1]
interceptions_df = interceptions_df.loc[interceptions_df['intNull']==0]
interceptions_df = interceptions_df[['playId','playerId']]
interceptions_df
```

```python
plays_df = pd.read_csv(input_file + '/' + plays_file)
plays_df = plays_df[['playId','gameId']]
plays_df
```

```python
plays_x_game = pd.merge(plays_df,games_df,how='inner',on='gameId')
plays_x_game
```

```python
interception_x_play = pd.merge(interceptions_df,plays_x_game,how='inner',on='playId')
interception_x_play
```

```python
interception_x_play.to_csv(output_file,index=False,header=True)
```
