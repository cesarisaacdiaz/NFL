```python
#Script for cleaning combine.csv file inside reosurces folder
#Developed by Cuitlahuac Daniel Maldonado Ruiz

#Importing dependencies
import os
import pandas as pd
import numpy as np
```

```python
#Files path
inputFile = os.path.join("..", "Resources", "combine.csv")
outputFile = os.path.join("..", "Clean_Data", "combine.csv")
```

```python
#Reading csv file into dataframe
combine_df = pd.read_csv(inputFile)

#Narrowing dataframe for players combineYear >= 2010
combine_df = combine_df.loc[combine_df["combineYear"] >= 2010]
combine_df
```

```python
#Creating a new dataframe only with needed columns
combine_df = combine_df[["playerId", "nameFull", "position", "collegeId", "college", "heightInches", "weight", "ageAtDraft"]]
combine_df

#Cleaning blank and NaN data
narrowed_df = combine_df.replace(r'^\s*$', np.NaN, regex=True)
narrowed_df.dropna(how="any", inplace=True)
narrowed_df
```

```python
#Formatting datatypes of columns
narrowed_df = narrowed_df.astype({"collegeId":"int64"})
narrowed_df = narrowed_df.astype({"ageAtDraft":"int32"})
narrowed_df
```

```python
#Wrtiting new csv file with cleaned and narrowed data
narrowed_df.to_csv(outputFile, index=False, header=True)
```
