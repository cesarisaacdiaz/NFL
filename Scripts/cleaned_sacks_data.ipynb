```python
#Dependencies
import pandas as pd
```

```python
# File to Load 
file_to_load = "../../Resources/sacks.csv"

# Read Purchasing File and store into Pandas data frame
sacks_data = pd.read_csv(file_to_load)
sacks_data
```

```python
#Cleaning for empty values
cleaned_data = sacks_data.dropna(axis=0)

#Remove useless columns
narrow_data = cleaned_data.drop(columns=['sackId','sackEnd'])

#Filter for specific value
narrow_data = narrow_data[narrow_data['sackType']==1]

#Filter for specific value
narrow_data = narrow_data[narrow_data['sackNull']==0]

#Re-format column
narrow_data = narrow_data.astype({"sackType":"int64"})

narrow_data
```

```python
narrow_data.to_csv("../Clean_Data/sacks.csv", index=False, header=True)
```
