```python
#Script for cleaning games.csv file inside reosurces folder
#Developed by Cesar Isaac Diaz Crespo

#importing dependencies
import os
import pandas as pd
import numpy as np
```

```python
#Files Path
input_file = os.path.join("..","Resources","games.csv")
output_file = os.path.join("..","Clean_Data","games.csv")
```

```python
#Reading csv file into a dataframe
games_df = pd.read_csv(input_file)

games_df = games_df.loc[games_df["gameDate"]>='2010']
games_df = games_df[games_df.seasonType.isin(['REG','POST','SB'])]
#games_df['seasonType'].value_counts()
games_df.head()
```

```python
#Creating a new DataFrame only with needed columns
games_df = games_df[["gameId","season","gameDate","homeTeamId","visitorTeamId","seasonType","winningTeam"]]
games_df.head()
```

```python
#Cleaning the data
games_df = games_df.astype({"gameDate":"str"})
games_df = games_df.astype({"seasonType":"str"})
games_df
```

```python
#Sending the data to csv file
games_df.to_csv(output_file,index=False,header=True)
```
